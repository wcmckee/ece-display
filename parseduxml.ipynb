script to parse rss xml feed from http://www.education.govt.nz/

```python
import requests
import xmltodict
```

```python
reqedu = requests.get('http://www.education.govt.nz/rss.xml')
```

```python
diedu = xmltodict.parse(reqedu.text)
```

```python
ledi = len(diedu['rss']['channel']['item'])
```
